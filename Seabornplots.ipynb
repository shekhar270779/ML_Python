```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Scatter plot
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', size='size', data=tips)
plt.show()
```

```python
dots = sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.relplot(x='time', y='firing_rate', col='align', hue='choice', kind='line', 
            legend='full', style='choice', size='coherence', 
            facet_kws=dict(sharex=False), data=dots)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(x='timepoint', y='signal', col='region', hue='event', style='event', kind='line', data=fmri)
```

```python
# linear regression model
sns.lmplot(x='total_bill', y='tip', col='time', hue='smoker', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='swarm', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='violin',  split=True,data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='bar', data=tips)
```

```python
f, axs = plt.subplots(1, 2, sharey=True, figsize=(6,4))
sns.boxplot(x='day', y='tip', data=tips, ax=axs[0])
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips, ax=axs[1])
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.jointplot(x='sepal_length', y='petal_length',  data=iris)
```

```python
sns.set(style='ticks', palette='muted')
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', col='time', hue='size', style='smoker', size='size',
            palette='YlGnBu', markers=['D','o'], sizes=(10,125),
            edgecolor='.2', linewidth=.5, alpha=.75, data=tips)
```

```python
g = sns.catplot(x='total_bill', y='day', hue='time', height=3.5, aspect=1.5,
               kind='box', legend=False, data=tips)
g.add_legend(title='Meal')
g.set_axis_labels('Total bill ($)','')
g.set(xlim=(0,60), yticklabels=['Thu','Fri','Sat','Sun'])
g.despine(trim=True)
g.fig.set_size_inches(6.5, 3.5)
g.ax.set_xticks([5,15,25,35,45,55], minor=True)
plt.setp(g.ax.get_yticklabels(),rotation=30 )

plt.show()
```

<a ref='https://seaborn.pydata.org/tutorial/axis_grids.html'><b>multi plot grids</b></a>

```python
g = sns.FacetGrid(data=tips, col='time')
g.map(plt.hist, 'tip')
```

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=0.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time', margin_titles=True)
g.map(sns.regplot, 'size', 'total_bill', color='.3', fit_reg=False, x_jitter=.1)
```

```python
ordered_categories = tips['day'].value_counts().index
```

```python
ordered_categories
```

```python
g = sns.FacetGrid(tips, row='day', row_order=ordered_categories, height=1.7, aspect=4)
g.map(sns.distplot, 'total_bill', hist=False, rug=True)
```

```python
pal = dict(Lunch='seagreen', Dinner='gray')
g = sns.FacetGrid(tips, hue='time', palette=pal, height=5)
g.map(plt.scatter, 'total_bill', 'tip', s=50, alpha=0.7, linewidth=.5, edgecolor='white')
g.add_legend()
```

```python
g = sns.FacetGrid(tips, hue='sex', palette='Set1', height=5, hue_kws={'marker':['^','v']})
g.map(plt.scatter, 'total_bill', 'tip', s=100, linewidth=0.5, edgecolor='white')
g.add_legend()
```
