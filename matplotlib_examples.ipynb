```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('classic')
```

```python
import numpy as np
```

```python
# Matlab Style

plt.figure(facecolor=(1,0,0,.1), figsize=(4,6))

plt.subplot(2,1,1)
x = np.arange(1, 10)
plt.plot(x, x**2)

plt.subplot(2,1,2)
plt.plot(x, x**3)

plt.show()
```

```python
# Object Oriented Style
fig, ax = plt.subplots(2, figsize=(4,6))

ax[0].plot(x, x**2)

ax[1].plot(x, x**3)
```

```python
x = np.array([2,3,4,5,6])

fig, ax = plt.subplots(figsize=(4,4))
ax.plot(x, x, Label='Linear')
ax.plot(x, x**2, Label='Square')
ax.plot(x, x**3, Label='Cubic')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Simple Plot')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, x**2, marker='*', color='b')
ax2.plot(x, x**3, marker='o', color='red')
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)

ax.plot([1,2,3,4], [10,15,22,30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], marker='^')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0,0].set_title('Upper Left')
ax[0,1].set_title('Upper Right')
ax[1,0].set_title('Lower Left')
ax[1,1].set_title('Lower Right')

for axes in ax.flat:
    axes.set(xticks=[], yticks=[])
```
