```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# How to initialize figure and axes
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

```python
# Create Simple plots 
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots() # creates a figure and an axes

ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Simple Plots')
ax.legend()
```

<h3> Simple bar chart </h3>

```python
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
men_std = [2,3,4,1,2]

fig, ax = plt.subplots()
ax.bar(groups, men_means, label='Men')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(groups, men_means, yerr=men_std, label='Men')
plt.show()
```

<h3>Stacked bar chart</h3>

```python
fig, ax = plt.subplots()
women_means = [25, 32, 34, 20, 25]
women_std = [3,5,2,3,3]

ax.bar(groups, men_means, yerr=men_std, label='Men')
ax.bar(groups, women_means, yerr=women_std, bottom=men_means, label='Women')
ax.set_xlabel('Groups')
ax.set_ylabel('Scores')
ax.set_title('Scores by Men and Women in different groups')
ax.legend()
plt.show()
```

<h3>Grouped bar chart</h3>

```python
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(groups))
width=0.35

fig, ax = plt.subplots()
rect1 = ax.bar(x - width/2, men_means, width, label='Men' )
rect2 = ax.bar(x +  width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by men/women in different groups')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                   xy=(rect.get_x() + rect.get_width()/2, height),
                   xytext=(0,3),
                   textcoords='offset points',
                   ha='center', va='bottom')
        
autolabel(rect1)        
autolabel(rect2)

fig.tight_layout()
plt.show()
```

<h3>Horizontal barchart</h3>

```python
np.random.seed(19680801)

plt.rcdefaults()

fig, ax = plt.subplots()

people = ['Tom', 'Dick', 'Harry', 'Jane', 'Jim']
y_pos = np.arange(len(people))
performance = [round(k,2) for k in 3 + 10 * np.random.rand(len(people))]
for x,y in zip(people, performance):
    print(x,y)
    
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()
ax.set_xlabel('Performance')
ax.set_title('Scores by Team')

plt.show()
```

<h3>Plotting Categorical Variables</h3>

```python
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,3, figsize=(9,3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

#axs[0].set_title('bar plot')
#axs[1].set_title('scatter plot')
#axs[2].set_title('line plot')

fig.suptitle('Categorical Plots')

plt.show()
```
