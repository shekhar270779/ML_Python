```python
import numpy as np
```

```python
t = np.arange(0, 5, 0.2)
t
```

```python
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.xlabel('X')
plt.ylabel('Exponent')
plt.show()
```

```python
plt.plot([1,2,3,4], [2,5,9,16], linewidth=5.0)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [2,3.5,7,9]

x2 = [1,2,3,4]
y2 = [3,7,10,15]

lines = plt.plot(x1,y1, x2,y2)

plt.setp(lines[0], color='r', linewidth=2.0, label='2019-Q')
plt.setp(lines[1], color='g', linewidth=3.0, label='2020-Q')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure(num=1)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
